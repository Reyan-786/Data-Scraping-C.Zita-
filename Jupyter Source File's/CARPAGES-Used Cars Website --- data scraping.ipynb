```python
from bs4 import BeautifulSoup
import requests 
import lxml
import pandas as pd
import numpy 
import os
```

```python
url='https://www.carpages.ca/used-cars/search/?fueltype_id%5B0%5D=3&fueltype_id%5B1%5D=7'
page=requests.get(url)
page
```

```python
soup=BeautifulSoup(page.text,'lxml')
soup
```

```python
postings = soup.find_all('div',class_='media soft push-none rule')
```

```python
df=pd.DataFrame({'title':[''],'link':[''],'price':[''],'location':[''],'km_driven':[''],'description and feautures':[''],'dealer info':['']})
i=0
while i<=15 :
    postings = soup.find_all('div',class_='media soft push-none rule')
    for post in postings:
        try:
            title = post.find('a').get('title')
            link = post.find('a').get('href')
            link_full='https://www.carpages.ca/'+link
            price= post.find('strong',class_='delta').text
            price=price.strip()
            location=post.find('p',class_='hN').text
            km_driven_status = post.find('div',class_='grey l-column l-column--small-6 l-column--medium-4').text
            km_driven = km_driven_status.strip()
            description_and_features = post.find('hgroup',class_='push-half--bottom').text
            description_and_features = description_and_features.strip()
            seller_info=post.find('hgroup',class_='vehicle__card--dealerInfo').text.strip()
            seller_info= seller_info[:20]
            length = len(df)
            df.loc[length]= {'title':title,'link':link_full,'price':price,'location':location,'km_driven':km_driven,'description and feautures':description_and_features,'dealer info':seller_info}
            print(df)
        except:
            pass
    next_page=soup.find('a',class_='nextprev').get('href')
    next_page_full='https://www.carpages.ca'+next_page
    url = next_page_full
    page = requests.get(url)
    soup=BeautifulSoup(page.text,'lxml')
    i=i+1
    
        
```
