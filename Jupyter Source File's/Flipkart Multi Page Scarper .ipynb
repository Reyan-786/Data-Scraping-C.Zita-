```python
from bs4 import BeautifulSoup 
import lxml
import pandas as pd
import requests
```

```python
url='https://www.flipkart.com/search?q=mobiles+&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off&page=1'
```

```python
page = requests.get(url)
```

```python
soup=BeautifulSoup(page.text,'lxml')
```

```python
df=pd.DataFrame({'Product Name':[''],'Link':[''],'Price':[''],'Rating':[''],'Description':['']})
i=0
while i<15:
    products = soup.find_all('div',class_='_1AtVbE col-12-12')
    for posts in products:
        try:
            title = posts.find('div',class_='_4rR01T').text
            link=posts.find('a',class_='_1fQZEK').get('href')
            link_full='https://www.flipkart.com'+link
            price=posts.find('div',class_='_30jeq3 _1_WHN1').text
            rating = posts.find('div',class_='_3LWZlK').text
            description= posts.find('li',class_='rgWa7D').text
            length=len(df)
            df.loc[length]={'Product Name':title,'Link':link_full,'Price':price,'Rating':rating,'Description':description}
            print(df)
        except:    
            pass

    next_page = soup.find('a',class_='_1LKTO3').get('href')
    next_page_full='https://www.flipkart.com'+next_page
    url=next_page_full
    page= requests.get(url)
    soup=BeautifulSoup(page.text,'lxml')
    i+=1


```
