```python
import lxml
from bs4 import BeautifulSoup
import pandas as pd 
import requests 
```

```python
url='https://www.nfl.com/standings/league/2019/REG'
```

```python
page=requests.get(url)
```

```python
soup=BeautifulSoup(page.text,'lxml')
```

```python
table = soup.find('table',{'summary':'Standings - Detailed View'})
```

```python
teamshtml = table.find_all('div',class_='d3-o-club-shortname')
```

```python
teams=[]
for i in teamshtml:
    name=i.text
    teams.append(name)
```

```python
df=pd.DataFrame(index =[teams ])
```

```python
df
```

```python
columns=[]
headings= table.find('tr')
```

```python
head_name = headings.find_all('tr',class_='header')
```

```python
headings = headings.find_all('th')
```

```python
for i in headings:
    name = i.text
    columns.append(name )
```

```python
df = pd.DataFrame(columns =[columns],index =[teams
                                            ])
```

```python
df
```

```python
table_rows= table.find_all('tr')[1:]
```

```python
teamstats=[]
for i in table_rows:
    row = i.find_all('td')
    stats = [j.text for j in row]
    teamstats.append(stats)
    
    
        
    
```

```python
df=pd.DataFrame(teamstats,columns=[columns],index=None)
df.reset_index(drop=True)
# df[1:]
df
```

```python
df2 = pd.DataFrame(columns=columns)
for row in table.find_all('tr')[1:]:
    data =row.find_all('td')
    row_data = [td.text for td in data]
    length = len(df2)
    df2.loc[length]=row_data
    
```
